```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
from matplotlib import rcParams

# Set font to Arial
rcParams['font.family'] = 'Arial'

# CIFAR-10 classes
classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')

# Get a colormap and extract a range of colors
colormap = plt.get_cmap('jet')
colors = colormap(np.linspace(0, 1, len(classes)))

# Create a plot
fig, ax = plt.subplots(figsize=(3, 5))
ax.axis('off')  # Turn off the axis

# Plot the class names with the corresponding colors
for i, class_name in enumerate(classes):
    ax.text(0.5, 1 - i * 0.06, class_name, fontsize=15, color=colors[i], ha='left', va='center', transform=ax.transAxes)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the pie chart
labels = [f'Class {i}' for i in range(1, 21)]
sizes = np.random.rand(20)
sizes = sizes / sizes.sum()  # Normalize to sum to 1

# Colors for the pie chart using the 'jet' colormap
colors = plt.cm.jet(np.linspace(0, 1, 20))

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, colors=colors, startangle=140)
plt.show()
```
