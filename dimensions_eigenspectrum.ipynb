```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib import rcParams
import seaborn as sns

# Set font to Arial
rcParams['font.family'] = 'Arial'

# Create the Set1 colormap
rainbow = sns.color_palette("magma", 4)


fig, (ax1) = plt.subplots(1, 1, figsize=(5, 4))

# First subplot
lambda0 = 30  # First eigenvalue
alpha = 0.02  # Decay constant
numEigenvalues = 90  # Number of eigenvalues to generate

i = np.arange(numEigenvalues)
eigenvalues = lambda0 * np.exp(-alpha * i)
y = 1000 * eigenvalues / np.sum(eigenvalues)
x = np.arange(1, numEigenvalues + 1) + 10

ax1.fill_between(x, 0, y, color=rainbow[0], alpha=0.5)

lambda0 = 30  # First eigenvalue
alpha = 0.1  # Decay constant
numEigenvalues = 90  # Number of eigenvalues to generate

i = np.arange(numEigenvalues)
eigenvalues = lambda0 * np.exp(-alpha * i)
y = 1000 * eigenvalues / np.sum(eigenvalues)
x = np.arange(1, numEigenvalues + 1) + 10

ax1.fill_between(x, 0, y, color=rainbow[3], alpha=0.5)

ax1.set_xlabel('Principal component', fontname="Arial", fontsize=20)
ax1.set_ylabel('Variance explained (%)', fontname="Arial", fontsize=20)
#ax.set_title('Activations in Layer 1', fontname="Arial", fontsize=25, fontweight="bold")

ax1.set_xticks([11, 50, 100])
ax1.set_xticklabels([1, 5, 10])
ax1.set_xlim([11, 100])
ax1.set_ylim([0, 100])
ax1.tick_params(axis='both', which='major', labelsize=14)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)

# Set the font name for axis tick labels to be Arial
for tick in ax1.get_xticklabels():
    tick.set_fontname("arial")
    tick.set_fontsize(20)
for tick in ax1.get_yticklabels():
    tick.set_fontname("arial")
    tick.set_fontsize(20)


plt.show()
```
